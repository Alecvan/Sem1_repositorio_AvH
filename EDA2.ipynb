### Outliers


```python
df['BMI'].describe()
```

```python
Q3 = np.percentile(df["BMI"], 80, interpolation = 'midpoint')
Q1 = np.percentile(df["BMI"], 20, interpolation = 'midpoint')
RIC = Q3 - Q1
df = df[(df['BMI'] <= Q3 + 1.5 * RIC) & (df['BMI'] >= Q1 - 1.5 * RIC)]
```
