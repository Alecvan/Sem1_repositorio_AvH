```python
brfss_reduced['FRIED_POTATOES'] = np.where((brfss_reduced['FRIED_POTATOES'] >= 100) & (brfss_reduced['FRIED_POTATOES'] <= 300), 1, brfss_reduced['FRIED_POTATOES'])
brfss_reduced['FRIED_POTATOES'] = np.where((brfss_reduced['FRIED_POTATOES'] >= 300) & (brfss_reduced['FRIED_POTATOES'] <= 400), 2, brfss_reduced['FRIED_POTATOES'])
brfss_reduced['FRIED_POTATOES'] = np.where((brfss_reduced['FRIED_POTATOES'] == 555), 3, brfss_reduced['FRIED_POTATOES'])
```

```python
brfss_reduced['FRIED_POTATOES'] = brfss_reduced['FRIED_POTATOES'].replace({777: '', 999: ''})
```
